```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import random
import re
```

```python
data_formatted = pd.read_pickle("data_formatted.pkl")
```

```python
classifier = pd.read_pickle("classifier_trained.pickle")
```

```python
data_formatted['text'] = data_formatted['text'].str.replace('@[^\s]+','')
data_formatted['text'] = data_formatted['text'].str.replace('[^\w\s]','')
data_formatted['text'] = data_formatted['text'].str.replace('\[.*?\]','')
data_formatted['text'] = data_formatted['text'].str.replace('[‘’“”…]','')
data_formatted['text'] = data_formatted['text'].str.replace('\w*\d\w*','')
```

```python
def tweet_text_split(tweet):
    tweet_words_in_list = [e.lower() for e in tweet.split()]
    tweet_words_in_list_wo_stopwords = [e for e in tweet_words_in_list if not e in stopwords.words('english')]
    return tweet_words_in_list_wo_stopwords
```

```python
data_formatted["tweet_text_split"] = data_formatted["text"].apply(tweet_text_split)
```

```python
def extract_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features[word] = (word in document_words)
    return features
```

```python
def get_words_in_tweets(tweets):
    all_words= []
    for words in tweets:
        all_words.extend(words)
    return all_words

def get_word_features(wordlist):
    wordlist = nltk.FreqDist(wordlist)
    word_features = list(wordlist.keys())
    return word_features

word_features = get_word_features(get_words_in_tweets(data_formatted["tweet_text_split"]))
```

```python
def join_words(wordlist):
    new_string = ' '.join(wordlist)
    return new_string
```

```python
data_formatted['Polarity Score'] = data_formatted["String New Tweet"].apply(lambda x: classifier.classify(extract_features(x.split())))
```

```python
data_formatted["String New Tweet"] = data_formatted["tweet_text_split"].apply(join_words)
```
