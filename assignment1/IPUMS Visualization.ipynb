```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ipums_proportion_data = pd.read_csv("Proportions_Ipums.csv")
```

```python
ipums_proportion_data
```

```python
ipums_proportion_data.sort_values("Proportion Some College").plot(kind = 'bar', x="State", y="Proportion Some College", 
                                                                  figsize=(15,5), legend = False, 
                                                                  ylim = (.25, .65), 
                                                                  title = "Proportion of Individuals with at Least One Year of College by State")
plt.xlabel("State")
plt.ylabel("Proportion of at Least One Year of College")
plt.show()
```

```python
ipums_proportion_data.sort_values("Proportion Unemployed").plot(kind = 'bar', x="State", y="Proportion Unemployed", 
                                                                figsize=(15,5), legend = False, 
                                                                ylim = (0.01, 0.06), 
                                                                title = "Proportion of Individuals Unemployed")
plt.xlabel("State")
plt.ylabel("Proportion Unemployed")
plt.show()
```
